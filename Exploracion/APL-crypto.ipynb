```python
import requests
import pandas as pd
url = 'https://api.coingecko.com/api/v3/'
headers = {
    "accept": "application/json",
    "x-cg-demo-api-key": "CG-key"
    }

endp_topten = 'coins/markets?vs_currency=usd&order=market_cap_desc&per_page=10&page=1'
```

```python
response = requests.get(url+'coins/list', headers=headers)
response_coin =  response.json()
df_coin = pd.json_normalize(response_coin)
```
